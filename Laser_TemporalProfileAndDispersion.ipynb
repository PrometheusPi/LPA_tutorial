```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from scipy.signal import hilbert, chirp
import scipy.constants as const
```

```python
# define maximum and minim wavelength 
lambda_max = 1000e-9 
lambda_min = 600e-9

# define laser central frequency
lambda_0 = 800e-9
```

```python
def wavelength_to_omega(wavelength):
    """
    convert light wavelength to (angular) frequency
    """
    return 2.0 * np.pi * const.c / wavelength
```

```python
N_freq = 1024
laser_frequency = np.linspace(wavelength_to_omega(lambda_max), 
                              wavelength_to_omega(lambda_min),
                              N_freq)
```

```python
delta_lambda = 20.0e-9

delta_omega = wavelength_to_omega(lambda_0 - delta_lambda) - wavelength_to_omega(lambda_0 + delta_lambda)
```

```python
laser_spectrum = np.exp(-0.5 * (laser_frequency - wavelength_to_omega(lambda_0))**2 / delta_omega**2)
```

```python
plt.plot(laser_frequency, laser_spectrum)
```

```python
plt.plot(np.fft.fftshift(np.real(np.fft.fft(laser_spectrum))))
plt.xlim(500, 540)
```
